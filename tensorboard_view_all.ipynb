```python
import argparse
import os
from random import randint
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

import numpy as np
import torch
import torchvision
from scipy.misc import imsave
from torch.utils.data import DataLoader

from torch.utils.tensorboard import SummaryWriter
from torchvision import models
from datetime import datetime
```

```python
parser = argparse.ArgumentParser()
parser.add_argument('--which_data', default='train', choices=['train', 'val'])
parser.add_argument('--which_model', default='M2', choices=['M1', 'M2'])
parser.add_argument('--M1_DIR', default='/home/dell/Documents/sg2im-master/output_samples/result_mix/')
parser.add_argument('--M2_DIR', default='/home/dell/Documents/scene_generation-master/output_samples/result_mix/')
```

```python
def main(args):
    now = datetime.now()
    current_time = now.strftime("%H_%M_%S")  
    model_name = args.which_model +'_'+ args.which_data
    RUN_DIR = os.path.join('tb_all_models', str(model_name), current_time)
    print(RUN_DIR)
    writer = SummaryWriter(RUN_DIR)
    
    
    if(args.which_model == 'M1'):
        directory = args.M1_DIR + args.which_data
    else:
        directory = args.M2_DIR + args.which_data
    print(directory)

    per = {'a':'5', 'b':'10', 'c':'20', 'd':'50', 'e':'100'}
    key = list(per.keys())
    
    for i in range(len(key)):
        file_names = os.listdir(directory)      
        selected_files =list(filter(lambda x: x.endswith(key[i]+'.png'), file_names))
        step = 0
        for file_name in sorted(selected_files):           
            label = args.which_model + args.which_data + '/'+per[key[i]]
            file_dir = os.path.join(directory, file_name)
            print(file_dir)
            img = Image.open(file_dir)
            imgT = torchvision.transforms.ToTensor()(img)
            writer.add_image(label, imgT, step)
            step += 1
   
```

```python
if __name__ == '__main__':
  args, _ = parser.parse_known_args()
  main(args)
```
